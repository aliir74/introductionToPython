````python
m1 = int(input())
m2 = int(input())
if(m2 > m1):
    m1, m2 = m2, m1

for i in range(8):
    a = int(input())
    if(a >= m1):
        m2 = m1
        m1 = a
    elif(a >= m2):
        m2 = a
    print('m1: ',m1, 'm2: ', m2)
        
print('max1: ', m1, 'max2: ', m2)
````
